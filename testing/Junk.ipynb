```python
import pandas as pd
```

```python
d19=pd.read_csv('2019-20.csv').set_index('full_name')
d18=pd.read_csv('2018-19.csv').set_index('full_name')
```

```python
for n in d18.index:
    if n not in d19.index:
        d18.drop(n,0,inplace=True)
        
        
    
```

```python
for n in d19.index:
    if n not in d18.index:
        d19.drop(n,0,inplace=True)
        
        
```

```python
d18.sort_values('full_name',inplace=True)
d19.sort_values('full_name',inplace=True)
```

```python
for i in d19:
    d19.rename(columns={i:(i+'_19')},inplace=True)
for i in d18:
    d18.rename(columns={i:(i+'_18')},inplace=True)
```

```python
df=pd.merge(d18,d19,on='full_name')
```
